## Introduction
WeRateDogs is a Twitter account that rates pictures of people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators, though? Almost always greater than 10. 11/10, 12/10, 13/10, etc. Why? Because "they're good dogs Brent."

It was started in 2015 by Matt Nelson, and has received international media coverage both for its popularity and for the attention drawn to social media copyright law when it was suspended by Twitter. As of October 2017, the Twitter account has close to four million followers and Nelson sees 30,000 likes on a post as being viral. His most popular post was of a dog marching in the 2017 Women's March, which was retweeted more than 50,000 times and favourited 134,000 times


```python
# Import the reqired packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline 
```

```python
# Read the data into a dataframe
twitter = pd.read_csv('twitter_archive_master.csv')
```

## Dog Ratings

```python
# Plot the ratings that are less than 20
plt.hist(twitter.rating_numerator[twitter.rating_numerator <= 20], 
         bins=np.arange(0,20,step=1), align='left');
plt.xticks(np.arange(0,20,step=1));
plt.title('Histogram of Ratings');
plt.xlabel('Rating (out of 10)');
plt.ylabel('Count');
```

- The most common ratings for dogs are 10/10, 11/10, 12/10, and 13/10, with the median rating being 11/10.
- An attribute of the rating system is the ability to rate multiple dogs at once. 
 * Let’s say that you have a picture of two dogs, and you want to rate them as 12/10. Since there are two of them, the rating would actually be 24/20! 
- With respect to the histogram of ratings, multi-dog ratings have already been normalized to be single-dog ratings.
- The vast majority of tweets were for a single dog.

### Favorites and Retweets
